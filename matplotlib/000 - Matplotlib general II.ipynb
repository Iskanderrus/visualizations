```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import random
from matplotlib import dates as dt
%matplotlib inline
%reload_ext autoreload
%autoreload 2
sns.set_theme()
```

# Functional plot

```python
x_1 = np.linspace(0, 10, 50)
y_1 = x_1 **2
```

```python
plt.plot(x_1, y_1)
plt.title("Squared numbers")
plt.xlabel("Numbers")
plt.ylabel("Squared numbers")
plt.show()
```

```python
# Creating equally spaced 100 data in range 0 to 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

# Generating x and y data
x = 16 * (np.sin(theta)**3)
y = 13 * np.cos(theta) - 5 * np.cos(2*theta) - 2 * \
    np.cos(3*theta) - np.cos(4*theta)

# Plotting
plt.plot(x, y, color='purple')
plt.title('Heart Shape')
plt.show()
```

## Printing multiple plots

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'or')
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'ob')
plt.show()
```

## Using figure objects

```python
fig_1 = plt.figure(figsize=(6, 8), dpi=200)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Numbers')
axes_1.set_ylabel('Squared numbers')
axes_1.set_title('Numbers Squared')
axes_1.plot(x_1, y_1, label='x/x**2')
axes_1.plot(y_1, x_1, label='x**2/x')
# --> location of legend: 
# 0-best, 1-upper right, 2-upper left, 3-lower left, 4-lower right, or tuple with (x,y) coordinates
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.45, 0.3])
axes_2.grid(False)
axes_2.set_xlabel('Numbers')
axes_2.set_ylabel('Squared numbers')
axes_2.set_title('Numbers Squared')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(0, 40, 'Message')
plt.show()
```

## SubPlots

```python
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharey=True)
plt.tight_layout()
axes_2[0].set_title('Squared numbers')
axes_2[0].set_xlabel('Number')
axes_2[0].set_ylabel('Number squared')
axes_2[1].set_title('Cubed numbers')
axes_2[1].set_xlabel('Number')
axes_2[1].set_ylabel('Number cubed')
axes_2[2].set_title('Combined plot')
axes_2[2].set_xlabel('Number')
axes_2[2].set_ylabel('Cubed and Squared combined')
axes_2[0].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[1].plot(x_1, x_1**3, 'r', alpha=0.7)
axes_2[2].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[2].plot(x_1, x_1**3, 'r', alpha=0.7)
plt.show()
```

```python
y_labels = ['Number squared', 'Numbers qubed', 'Cubed and Squared combined']
titles = ['Squared numbers', 'Cubed numbers', 'Combined plot']
colors = ['b', 'r', 'b', 'r']

fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharey=True)
plt.tight_layout()

for x in range(3): 
    axes_2[x].set_title(titles[x])
    axes_2[x].set_xlabel('Number')
    axes_2[x].set_ylabel(y_labels[x])

axes_2[0].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[1].plot(x_1, x_1**3, 'r', alpha=0.7)
axes_2[2].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[2].plot(x_1, x_1**3, 'r', alpha=0.7)

plt.show()
```

## Histograms

```python
arr_1 = np.random.randint(1, 7, 10250)
arr_2 = np.random.randint(1, 7, 10250)
arr_3 = arr_1 + arr_2
```

```python
density = stats.gaussian_kde(arr_3)

n, bins, _ = plt.hist(arr_3, bins=11, density=True, stacked=False, color='green', alpha=0.5)
plt.plot(bins, density(bins))
plt.show()
```

## Pie Charts

```python
fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
types = [
    'Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
    'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon',
    'Fairy', 'Ghost', 'Ice'
]
poke_num = random.sample(range(1, 150), len(types))
colors = []
for i in range(len(types)):
    rgb = (random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5))
    colors.append(rgb)

explode = [0.01] * len(types)
explode[0] = 0.5

wedges, texts, autotexts = plt.pie(poke_num,
                                   explode=explode,
                                   labels=types,
                                   colors=colors,
                                   autopct='%0.1f%%',
                                   shadow=True,
                                   startangle=30,
                                   textprops=dict(color='w'))

plt.legend(wedges, types, loc='best', bbox_to_anchor=(0.8, 0, 0.5, 1))

plt.show()
```
