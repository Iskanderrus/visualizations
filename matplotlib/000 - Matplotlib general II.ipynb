## Functional plot

```python
x_1 = np.linspace(0, 10, 50)
y_1 = x_1 **2
```

```python
plt.plot(x_1, y_1)
plt.title("Squared numbers")
plt.xlabel("Numbers")
plt.ylabel("Squared numbers")
plt.show()
```

```python
# Creating equally spaced 100 data in range 0 to 2*pi
theta = np.linspace(0, 2 * np.pi, 100)

# Generating x and y data
x = 16 * (np.sin(theta)**3)
y = 13 * np.cos(theta) - 5 * np.cos(2*theta) - 2 * \
    np.cos(3*theta) - np.cos(4*theta)

# Plotting
plt.plot(x, y, color='purple')
plt.title('Heart Shape')
plt.show()
```

## Printing multiple plots

```python
plt.subplot(1, 2, 1)
plt.plot(x_1, y_1, 'or')
plt.subplot(1, 2, 2)
plt.plot(x_1, y_1, 'ob')
plt.show()
```

## Using figure objects

```python
fig_1 = plt.figure(figsize=(6, 8), dpi=200)
axes_1 = fig_1.add_axes([0.1, 0.1, 0.9, 0.9])
axes_1.set_xlabel('Numbers')
axes_1.set_ylabel('Squared numbers')
axes_1.set_title('Numbers Squared')
axes_1.plot(x_1, y_1, label='x/x**2')
axes_1.plot(y_1, x_1, label='x**2/x')
# --> location of legend: 
# 0-best, 1-upper right, 2-upper left, 3-lower left, 4-lower right, or tuple with (x,y) coordinates
axes_1.legend(loc=0)

axes_2 = fig_1.add_axes([0.45, 0.45, 0.45, 0.3])
axes_2.grid(False)
axes_2.set_xlabel('Numbers')
axes_2.set_ylabel('Squared numbers')
axes_2.set_title('Numbers Squared')
axes_2.plot(x_1, y_1, 'r')
axes_2.text(0, 40, 'Message')
plt.show()
```

## SubPlots

```python
fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharey=True)
plt.tight_layout()
axes_2[0].set_title('Squared numbers')
axes_2[0].set_xlabel('Number')
axes_2[0].set_ylabel('Number squared')
axes_2[1].set_title('Cubed numbers')
axes_2[1].set_xlabel('Number')
axes_2[1].set_ylabel('Number cubed')
axes_2[2].set_title('Combined plot')
axes_2[2].set_xlabel('Number')
axes_2[2].set_ylabel('Cubed and Squared combined')
axes_2[0].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[1].plot(x_1, x_1**3, 'r', alpha=0.7)
axes_2[2].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[2].plot(x_1, x_1**3, 'r', alpha=0.7)
plt.show()
```

```python
y_labels = ['Number squared', 'Numbers qubed', 'Cubed and Squared combined']
titles = ['Squared numbers', 'Cubed numbers', 'Combined plot']
colors = ['b', 'r', 'b', 'r']

fig_2, axes_2 = plt.subplots(figsize=(8, 4), nrows=1, ncols=3, sharey=True)
plt.tight_layout()

for x in range(3): 
    axes_2[x].set_title(titles[x])
    axes_2[x].set_xlabel('Number')
    axes_2[x].set_ylabel(y_labels[x])

axes_2[0].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[1].plot(x_1, x_1**3, 'r', alpha=0.7)
axes_2[2].plot(x_1, x_1**2, 'b', alpha=0.5)
axes_2[2].plot(x_1, x_1**3, 'r', alpha=0.7)

plt.show()
```

## Pie Charts

```python
fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
types = [
    'Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
    'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon',
    'Fairy', 'Ghost', 'Ice'
]
poke_num = random.sample(range(1, 150), len(types))
colors = []
for i in range(len(types)):
    rgb = (random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5))
    colors.append(rgb)

explode = [0.01] * len(types)
explode[0] = 0.5

wedges, texts, autotexts = plt.pie(poke_num,
                                   explode=explode,
                                   labels=types,
                                   colors=colors,
                                   autopct='%0.1f%%',
                                   shadow=True,
                                   startangle=30,
                                   textprops=dict(color='w'))

plt.legend(wedges, types, loc='best', bbox_to_anchor=(0.8, 0, 0.5, 1))

plt.show()
```

## 3D Surfaces

```python
fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')

z = 15 * np.random.random(100)
x = np.sin(z) * np.random.randn(100)
y = np.sin(z) * np.random.randn(100)
ax.scatter3D(x, y, z, c=z, cmap='coolwarm')
plt.show()
```

```python
def get_x(x, y):
    return np.sin(np.sqrt(x**2 + y**2))


x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)
x, y = np.meshgrid(x, y)

z = get_x(x, y)

fig = plt.figure(figsize=(8, 5), dpi=200)
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9], projection='3d')
ax.view_init(45, 55)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
#ax.contour3D(x, y, z, 180, cmap='coolwarm')
#ax.plot_wireframe(x, y, z, color=['b'])
ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap='coolwarm', edgecolor='none')

plt.show()
```

```python
import mplfinance as mpf
```

```python
df = pd.read_csv('T.csv', index_col=0, parse_dates=True)
```

```python
df.head()
```

```python
df.index.name = 'Date'
```

```python
mpf.plot(df.iloc[100:195, :], type='candle')
```

```python
mpf.plot(df.tail(365), type='ohlc', mav=(3,5,7,15), volume=True, show_nontrading=True)
```

```python
mpf.plot(df.tail(365), type='pnf')
```

```python
data = df.tail(365)
data['MA20'] = data['Adj Close'].rolling(20).mean()
data['MA50'] = data['Adj Close'].rolling(50).mean()
data = data.dropna()

buy_signals = []
sell_signals = []

for i in range(len(data)):
    if (data['MA20'].iloc[i] > data['MA50'].iloc[i]) and (data['MA20'].iloc[i-1] < data['MA50'].iloc[i-1]):
        buy_signals.append(data.iloc[i]['Adj Close'] * 0.98)
    else:
        buy_signals.append(np.nan)
    
    if (data['MA20'].iloc[i] < data['MA50'].iloc[i]) and (data['MA20'].iloc[i-1] > data['MA50'].iloc[i-1]):
        sell_signals.append(data.iloc[i]['Adj Close'] * 1.02)
    else:
        sell_signals.append(np.nan)

buy_markers = mpf.make_addplot(buy_signals, type='scatter', markersize=120, marker='^')
sell_markers = mpf.make_addplot(sell_signals, type='scatter', markersize=120, marker='v')
apds = [buy_markers, sell_markers]
mpf.plot(data, type="candle", addplot=apds)
```
