```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Chart

```python
# data generation - 1000 evenly distributed samples in the range 0 to 20:  
x = np.linspace(0, 20, 1000)
```

```python
# y-axis data: 
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y); 
```

```python
plt.plot(x,y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Simple Line Plot')
plt.grid();
```
