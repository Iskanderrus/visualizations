```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Chart

```python
# data generation - 1000 evenly distributed samples in the range 0 to 20:  
x = np.linspace(0, 20, 1000)
```

```python
# y-axis data: 
y = np.sin(x) + 0.2*x
```

```python
plt.plot(x,y); 
```

```python
plt.plot(x,y)
plt.xlabel('Input')
plt.ylabel('Output')
plt.title('Simple Line Plot')
plt.grid();
```

# Scatterplot

```python
x = np.random.randn(100, 2)
```

```python
plt.scatter(x[:, 0], x[:, 1]); 
```

```python
x = np.random.randn(200, 2)
x[:50] += 3
y = np.zeros(200)
y[:50] = 1
plt.scatter(x[:, 0], x[:, 1], c=y);
```

# Histogram

```python
# standard distribution: 
x = np.random.randn(10000)
```

```python
plt.hist(x);
```

```python
plt.hist(x, bins=50);
```

```python
# uniform distribution: 
x = np.random.random(10000)
```

```python
plt.hist(x, bins=50);
```
